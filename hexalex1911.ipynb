```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
WORDS = [
    'AWHILE',
    'COD',
    'FUN',
    'JOT',
    'KEG',
    'KELP',
    'NAVY',
    'OHM',
    'RELIC',
    'RIOTS',
    'WAVY', 
    'WAX',
    'ZAP',
]

letters = set(''.join(WORDS))
#letters
```

```python

H = nx.Graph()
H.add_nodes_from([ltr for ltr in letters])


edges = [(word[i], word[i+1]) for word in WORDS for i in range(len(word)-1)]
print(edges)
H.add_edges_from(edges)

# Print the nodes and edges
print("Nodes:", H.nodes())
print("Edges:", H.edges())


nx.draw(H, with_labels=True, node_color='lightblue')
plt.show()
```
