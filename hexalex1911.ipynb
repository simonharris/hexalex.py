```python
import matplotlib.pyplot as plt
import networkx as nx
```

### Add each node

```python
WORDS = [
    'AWHILE',
    'COD',
    'FUN',
    'JOT',
    'KEG',
    'KELP',
    'NAVY',
    'OHM',
    'RELIC',
    'RIOTS',
    'WAVY', 
    'WAX',
    'ZAP',
]

letters = set(''.join(WORDS))

H = nx.Graph()
H.add_nodes_from([ltr for ltr in letters])

```

```python
PREPOPULATED = [
    'F',
    'S',
    'D',
    'V',
    'I',
    'Z',
    'K',
    'G',
]


# Add node attribute 'filled'
for node in H.nodes:
    H.nodes[node]['filled'] = node in PREPOPULATED
```

### Add grid positions

```python

# Define node positions
pos = {
    'F': (0, 7),
    'U': (0, 5),
    'N': (0, 3),
    'X': (0, 1),

    'Y': (1, 6),
    'V': (1, 4),
    'A': (1, 2),
    'Z': (1, 0),
    
    'S': (2, 7),
    'M': (2, 5),
    'W': (2, 3),
    'P': (2, 1),

    'T': (3, 6),
    'H': (3, 4),
    'L': (3, 2),
    'K': (3, 0),
     
    'D': (4, 7),
    'O': (4, 5),
    'I': (4, 3),
    'E': (4, 1),

    'J': (5, 6),
    'C': (5, 4),
    'R': (5, 2),
    'G': (5, 0),
}
```

### Add pre-populated nodes

```python



edges = [(word[i], word[i+1]) for word in WORDS for i in range(len(word)-1)]
print(edges)
H.add_edges_from(edges)

# Print the nodes and edges
print("Nodes:", H.nodes())
print("Edges:", H.edges())


nx.draw(H, with_labels=True, node_color='lightblue')
plt.show()

```

```python
# Draw the graph with custom node colors and labels
node_colors = ['lightblue' if H.nodes[node]['filled'] else 'pink' for node in H.nodes]
node_labels = {node: node if H.nodes[node]['filled'] else '' for node in H.nodes}

nx.draw(H, with_labels=True, labels=node_labels, node_color=node_colors)
plt.show()
```
